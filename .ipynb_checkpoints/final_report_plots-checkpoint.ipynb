```python
from IPython.display import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import xclim as xc
from operator import sub
import os as os
# Geographic ploting
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from shapely.geometry import Polygon
from cartopy.io import shapereader
import cartopy.io.img_tiles as cimgt
import geopandas
```

```python
def rect_from_bound(xmin, xmax, ymin, ymax):
    """Returns list of (x,y)'s for a rectangle"""
    xs = [xmax, xmin, xmin, xmax, xmax]
    ys = [ymax, ymax, ymin, ymin, ymax]
    return [(x, y) for x, y in zip(xs, ys)]

lon_min = 4
lon_max = 12
lat_min = 44
lat_max = 50

exts = [lon_min, lon_max, lat_min, lat_max]

# request data for use by geopandas
resolution = '10m'
category = 'cultural'
name = 'admin_0_countries'

shpfilename = shapereader.natural_earth(resolution, category, name)
df = geopandas.read_file(shpfilename)

# get geometry of a country
poly = [df.loc[df['ADMIN'] == 'Switzerland']['geometry'].values[0]]

stamen_terrain = cimgt.Stamen('terrain-background')

# projections that involved
st_proj = stamen_terrain.crs  #projection used by Stamen images
ll_proj = ccrs.PlateCarree()  #CRS for raw long/lat

# make a mask polygon by polygon's difference operation
# base polygon is a rectangle, another polygon is simplified switzerland
msk = Polygon(rect_from_bound(*exts)).difference( poly[0].simplify(0.01) )
msk_stm  = st_proj.project_geometry (msk, ll_proj)  # project geometry to the projection used by stamen
```

```python
chain_names = ["Mod 1; GCM: EC-EARTH; RCM:CLIM", "Mod 2; GCM: EC-EARTH; RCM: SMHI", "Mod 3; GCM: MPI; RCM:REMO", "Mod 4; GCM: MPI;RCM:SMHI"]
```

```python
ds_pop_rcp = xr.open_dataset("ds_pop_rcp.nc")
ds_precip_rcp = xr.open_dataset("ds_precip_rcp.nc")
ds_precip_hist = xr.open_dataset("ds_precip_hist.nc")
```

```python
rcp_pop = ds_pop_rcp.population/(12.5*12.5)
```

```python
[rcp_pop.values.min(), rcp_pop.values.max()]
```

```python
intens_precip = [ds_precip_rcp.isel(mod_chain=nr).quantile(q=.99, dim='time') - ds_precip_hist.isel(mod_chain=nr).quantile(q=.99, dim='time') for nr in [0,1,2,3]]
```

```python
#range intensity
np.array([[item.pr.values.min() for item in intens_precip],[item.pr.values.max() for item in intens_precip]])
```

```python
freq_precip = [xc.icclim.R95p(ds_precip_rcp.isel(mod_chain=nr).pr, ds_precip_hist.isel(mod_chain=0).pr.quantile(0.99, dim="time", keep_attrs=True)).mean(dim="time") for nr in [0,1,2,3]]
```

```python
#range frequency
np.array([[item.min() for item in freq_precip],[item.max() for item in freq_precip]])
```

```python
import seaborn as sns
```

```python
#INTENSITY PLOT
sns.set(context="paper", font_scale=2)

projection=ccrs.PlateCarree()
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(20,15),  dpi=100, 
                                         subplot_kw=dict(projection=ccrs.PlateCarree())
                                        )

intens_lev = [-14, -9, -4, -1, 1, 4, 9, 14]

#fig.suptitle('Difference in 0.99 percentile daily precipitation between 2071-2100 and 1971-2000',fontsize='large', fontweight='bold', y=0.95,x=0.5)
for ax, data, names in zip(axes.flat, intens_precip,chain_names):
    p = data.pr.plot.contourf(ax=ax, levels=intens_lev, cmap="RdBu_r", add_colorbar=False, extend="both")
    ax.set_aspect(aspect=1.35)
    ax.set_title('')
    ax.set_title(names,fontweight='bold')
    plt.subplots_adjust(wspace = 0.04, right=1, hspace=-0.4, top=1)
    ax.add_feature(cfeature.BORDERS)
    ax.add_geometries(msk_stm, st_proj, zorder=12, facecolor='white', edgecolor='black', alpha=1, linewidth=2)
plt.colorbar(p, ax=axes.flat, shrink=0.7, label="Difference in daily precipitation [mm/d]", ticks=intens_lev,
            orientation='vertical')   

plt.savefig('figures/intens.jpg', dpi=300, bbox_inches='tight', pad_inches=0.05)
```

```python
#FREQUENCY PLOT
sns.set(context="paper", font_scale=2)

projection=ccrs.PlateCarree()
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(20,15),  dpi=100, 
                                         subplot_kw=dict(projection=ccrs.PlateCarree())
                                        )

freq_lev = [0,2,5,10,15,20,30]

#fig.suptitle('Number of days between 2071-2100 exceeding the 0.99 percentile precipitation of 1971-2000',fontsize='large', fontweight='bold', y=0.95,x=0.5)
for ax, data, names in zip(axes.flat, freq_precip,chain_names):
    p = data.plot.contourf(ax=ax, vmin=0, vmax=5, levels=freq_lev, cmap="viridis", add_colorbar=False, extend="max")
    ax.set_aspect(aspect=1.35)
    ax.set_title('')
    ax.set_title(names,fontweight='bold')
    plt.subplots_adjust(wspace = 0.04, right=1, hspace=-0.4, top=1)
    ax.add_feature(cfeature.BORDERS)
    ax.add_geometries(msk_stm, st_proj, zorder=12, facecolor='white', edgecolor='black', alpha=1, linewidth=2)
plt.colorbar(p, ax=axes.flat, shrink=0.7, label="number of days ", ticks=freq_lev,
            orientation='vertical')   

plt.savefig('figures/freq.jpg', dpi=300, bbox_inches='tight', pad_inches=0.05)
```

```python
#POPULATION PLOT
sns.set(context="paper", font_scale=2)

fig = plt.figure(figsize=(14,10))
ax = plt.axes(projection=ccrs.PlateCarree())

pop_lev = [0,1000,2000,3000,4000,5000,6000]

fig.subplots_adjust(top=1, right=1)
a = rcp_pop.plot.contourf(ax=ax, transform=ccrs.PlateCarree(), add_colorbar=False,levels=pop_lev, extend="max")
ax.set_aspect(aspect=1.35)
cbar = fig.colorbar(a, ax=ax, fraction = 0.1, label=r'individuals / km$^2$', ticks=pop_lev, shrink=0.9, orientation="horizontal",pad=0.001)

ax.add_geometries(msk_stm, st_proj, zorder=12, facecolor='white', edgecolor='black', alpha=1, linewidth=2)

plt.savefig('figures/pop.jpg', dpi=300, bbox_inches='tight', pad_inches=0.05)
```
