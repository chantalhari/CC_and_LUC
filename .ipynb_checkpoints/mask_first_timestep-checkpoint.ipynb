```python
import xarray as xr

def remap_to_common_coords(source_da, target_da):
    return source_da.interp(coords={'latitude': target_da.lat, 'longitude': target_da.lon}, method='nearest')

# Open your data arrays
binary_mask = xr.open_dataarray('binary_mask.nc', decode_times=False)
recognized_grid_da = xr.open_dataset('first_timestep.nc', decode_times=False)

# Remap binary_mask to the grid of recognized_grid_da
remapped_binary_mask = remap_to_common_coords(binary_mask, recognized_grid_da)

# Save the remapped binary_mask
remapped_binary_mask.to_netcdf('remapped_binary_mask.nc')
```

```python
import xarray as xr

# Load the datasets
binary_mask_ds = xr.open_dataset('binary_mask.nc', decode_times=False)
remapped_current_timestep_ds = xr.open_dataset('remapped_current_timestep.nc', decode_times=False)

# Function to print grid size and resolution
def print_grid_info(ds, name):
    print(f"Grid Info for {name}:")
    if 'latitude' in ds.coords and 'longitude' in ds.coords:
        lat = ds.coords['latitude']
        lon = ds.coords['longitude']
        
        print(f"  Latitude: {lat.size} points, from {lat.min().item()} to {lat.max().item()}")
        print(f"  Longitude: {lon.size} points, from {lon.min().item()} to {lon.max().item()}")
        
        if lat.size > 1 and lon.size > 1:
            lat_res = abs(lat[1] - lat[0]).item()
            lon_res = abs(lon[1] - lon[0]).item()
            print(f"  Resolution: {lat_res} in latitude, {lon_res} in longitude")
    else:
        print("  Dimensions:")
        for dim, size in ds.dims.items():
            print(f"    {dim}: {size}")
        
# Print the grid information
print_grid_info(binary_mask_ds, 'binary_mask.nc')
print_grid_info(remapped_current_timestep_ds, 'remapped_current_timestep.nc')
```

```python
import xarray as xr

# Open the datasets
binary_mask_ds = xr.open_dataset('binary_mask.nc', decode_times=False)
remapped_current_timestep_ds = xr.open_dataset('remapped_current_timestep.nc', decode_times=False)


# You may need to extract the actual variable from the Dataset, 
# assuming here 'var_name' is the variable name in 'remapped_current_timestep_ds'

binary_mask_da = binary_mask_ds['wdpa_data']  # Replace with actual variable name of binary mask
# Initialize a new Dataset to hold the masked variables
masked_current_ds = xr.Dataset()

# Loop over all variables in remapped_current_timestep_ds and perform multiplication
for var_name in remapped_current_timestep_ds.data_vars:
    remapped_current_timestep_da = remapped_current_timestep_ds[var_name]
    masked_current_da = remapped_current_timestep_da * binary_mask_da
    
    # Assign the result back to the new Dataset
    masked_current_ds[var_name] = masked_current_da

# Save the result to a new NetCDF file
masked_current_ds.to_netcdf('masked_current.nc')
```
