```python
import os
import pandas as pd

taxas = ["Mammals"]
models = ["GAM", "GBM"]

# This will store the formatted names for each model
model_formatted_names = {}

for taxa in taxas:
    for model in models:
        convcodes = pd.read_csv("/storage/homefs/ch21o450/scripts/BioScenComb/data/IUCN_LUH_converion_table_Carlson.csv")
        dir_habclass = "/storage/homefs/ch21o450/IUCN/Habitat_Classifications/" + taxa + "/"
        dir_species = "/storage/workspaces/wa_climate/climate_trt/data/BioScen15/individual_projections/" + taxa + "_" + model + "_results_climate/"
        available_file = os.listdir(dir_species)
        available_names = [x.split(".csv")[0] for x in available_file]

        formatted_names = []
        for species_name in available_names:
            split_species_name = species_name.split("_")[:2]
            formatted_species_name = " ".join(split_species_name)
            formatted_names.append(formatted_species_name)
        
        # Store the formatted names for this model
        model_formatted_names[model] = formatted_names

# Now compare the formatted names for GAM and GBM
if set(model_formatted_names["GAM"]) == set(model_formatted_names["GBM"]):
    print("The formatted names for GAM and GBM are identical!")
else:
    unique_to_GAM = set(model_formatted_names["GAM"]) - set(model_formatted_names["GBM"])
    unique_to_GBM = set(model_formatted_names["GBM"]) - set(model_formatted_names["GAM"])

    if unique_to_GAM:
        print("Names unique to GAM:")
        for name in unique_to_GAM:
            print(name)

    if unique_to_GBM:
        print("Names unique to GBM:")
        for name in unique_to_GBM:
            print(name)
```

```python
import difflib

def find_differences(file1, file2):
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        content1 = f1.readlines()
        content2 = f2.readlines()

    # Check if sorted contents are the same
    if sorted(content1) == sorted(content2):
        return content1, content2, ["Same entries, but different order."]

    d = difflib.Differ()
    diff = list(d.compare(content1, content2))

    differences = [line for line in diff if line.startswith('- ') or line.startswith('+ ')]
    return content1, content2, differences

file_paths = [
    "/storage/homefs/ch21o450/scripts/BioScenComb/habitat_counts/habitat_pasture_Mammals_GAM_35.csv",
    "/storage/homefs/ch21o450/scripts/BioScenComb/habitat_counts/habitat_pasture_Mammals_GBM_35.csv",
        "/storage/homefs/ch21o450/scripts/BioScenComb/habitat_counts/habitat_pasture_Mammals_GAM_65.csv",
    "/storage/homefs/ch21o450/scripts/BioScenComb/habitat_counts/habitat_pasture_Mammals_GBM_65.csv"
]

files = []

for i in range(len(file_paths)):
    for j in range(i + 1, len(file_paths)):
        content_i, content_j, differences = find_differences(file_paths[i], file_paths[j])
        
        files.append(content_i)  # Appending the content of file i
        files.append(content_j)  # Appending the content of file j
        
        if differences:
            print(f"Differences between {file_paths[i]} and {file_paths[j]}:")
            for diff in differences:
                print(diff)
            print("-" * 40)
        else:
            print(f"{file_paths[i]} and {file_paths[j]} are identical.")
```
